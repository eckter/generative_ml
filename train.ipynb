```python
%load_ext autoreload
%autoreload 2

from pathlib import Path
import generative_ml
from generative_ml.models import VAE, DummyVAE
from generative_ml.training import train_vae
from torchvision import transforms
import torch
from IPython.display import display
```

```python
data_dir = Path() / "data"
```

```python
params = generative_ml.HyperParams(
    n_epoch=10_000,
    sample_every_n_step=200,
    learning_rate=5e-3
)

# Small dataset: we just check that we can overfit here
data = generative_ml.PokeDataset(data_dir, image_size=params.image_size, n_pokemon=8)

model = DummyVAE(params.latent_size, params.image_size)
for sample in train_vae(model, data, params):
    print(sample.n_step)
    display(transforms.ToPILImage()(torch.concatenate(list(sample.outputs[:6]), dim=2)))
```
