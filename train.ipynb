```python
%load_ext autoreload
%autoreload 2

from pathlib import Path
import generative_ml
from generative_ml.models.dummy_vae import DummyVAE
from generative_ml.training import train_vae
from torchvision import transforms
import torch
from IPython.display import display
```

```python
data_dir = Path() / "data"
```

```python
params = generative_ml.HyperParams(
    n_epoch=100_000,
    sample_every_n_step=200,
    learning_rate=5e-3
)
data = generative_ml.PokeDataset(data_dir, image_size=params.image_size, n_pokemon=32)
model = DummyVAE(params.latent_size, params.image_size)
for sample in train_vae(model, data, params):
    print(sample.n_step)
    #print("in:")
    #for x in sample.inputs:
    #    display(transforms.ToPILImage()(x))
    #print("out:")
    for x in sample.outputs[:1]:
        display(transforms.ToPILImage()(x))
```
